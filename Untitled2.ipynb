```python
import pandas as pd

# Cargar el archivo
df = pd.read_excel("ATLANTICO.xls")

# Ver las primeras filas para confirmar que se cargó bien
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter

# Función para formatear miles con K
def format_thousands(x, pos):
    return '%1.0fK' % (x*1e-3) if x >= 1000 else '%1.0f' % x

formatter = FuncFormatter(format_thousands)

# Etiquetas y datos de ejemplo (ajusta si los tomas del Excel)
edad_labels = ['0-4 años', '5-9 años', '10-14 años', '15-44 años', '45-59 años', '60 y más', 'Sin info']
datos_por_edad = {
    'Caminar/correr/saltar': [310, 391, 469, 2558, 1539, 3653, 12],
    'Desplazarse cortos trechos': [149, 206, 229, 1032, 1043, 2623, 3],
    'Mover objetos con manos': [94, 189, 186, 876, 461, 1143, 3],
    'Cambiar posiciones cuerpo': [112, 129, 160, 839, 461, 1200, 5]
}

# Gráfico 4: Motoras
fig4, ax4 = plt.subplots(figsize=(12, 7))

dificultades_motoras = list(datos_por_edad.keys())
colors_motoras = ['#E74C3C', '#F39C12', '#9B59B6', '#16A085']

bottom = np.zeros(len(edad_labels))
for i, dificultad in enumerate(dificultades_motoras):
    valores = datos_por_edad[dificultad]
    ax4.bar(edad_labels, valores, bottom=bottom, label=dificultad,
            color=colors_motoras[i], alpha=0.9, edgecolor='black', linewidth=0.5)
    bottom += valores

ax4.yaxis.set_major_formatter(formatter)
ax4.set_ylabel('Cantidad de Personas', fontweight='bold', fontsize=11)
ax4.set_xlabel('Grupos de Edad', fontweight='bold', fontsize=11)
ax4.set_title('DIFICULTADES MOTORAS POR GRUPOS DE EDAD',
              pad=20, fontweight='bold', fontsize=14, loc='center')
ax4.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

ax4.grid(axis='y', alpha=0.3, linestyle='--')
ax4.set_axisbelow(True)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()  # 👈 IMPORTANTE: Muestra la gráfica en Colab
```

```python
datos_por_edad = {
    'Caminar/correr/saltar': df['Caminar/correr/saltar'].tolist(),
    'Desplazarse cortos trechos': df['Desplazarse cortos trechos'].tolist(),
    'Mover objetos con manos': df['Mover objetos con manos'].tolist(),
    'Cambiar posiciones cuerpo': df['Cambiar posiciones cuerpo'].tolist()
}
```
